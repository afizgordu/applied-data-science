```python
df = pd.read_csv("yelp.csv")
```

```python
df.head()
```

```python
df.text[0]
```

```python
df['text']=df['text'].str.lower() #küçük harfe çevir
df['text']=df['text'].str.replace('[^\w\s]', ' ',regex=True) #noktalama işaretlerini kaldırır
df['text']=df['text'].str.replace('\n', ' ',regex=True) #satır sonlarını kaldırır
df['text']=df['text'].str.replace('\d+', ' ',regex=True) #rakamları kaldırır
df['text']=df['text'].str.replace('\r', ' ',regex=True) #satır başlarını baldırır
```

```python
ybw = df[(df.stars == 1)|(df.stars == 5)]
```

```python
ybw.reset_index(drop = True,inplace=True)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
```

```python
vect=CountVectorizer(stop_words='english',ngram_range=(1,2))
```

```python
x=ybw[['text']]
y=ybw[['stars']]
```

```python
xyeni=vect.fit_transform(x['text'])
```

```python
x_train,x_test,y_train,y_test=train_test_split(xyeni,y,random_state=42,test_size=0.20)
```

```python
nb=MultinomialNB()
```

```python
model=nb.fit(x_train,y_train)
```

```python
pred=model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,pred)
```
