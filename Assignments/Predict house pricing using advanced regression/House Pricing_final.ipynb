##### 2-handling with the Miscs

```python
# 1. Adım: 'MiscFeature' alanı dolu olup 'MiscVal' değeri sıfır olanları, 'MiscFeature' değerlerinin ortalaması ile doldur
mask1 = (df['MiscFeature'].notna()) & (df['MiscVal'] == 0)
mean_values = df[df['MiscVal'] != 0].groupby('MiscFeature')['MiscVal'].mean().astype(int)
df.loc[mask1, 'MiscVal'] = df.loc[mask1, 'MiscFeature'].map(mean_values)

# 2. Adım: 'MiscFeature' alanı boş olup 'MiscVal' değeri 0 olan kayıtları 'Nmf' ile doldur
mask2 = df['MiscFeature'].isna() & (df['MiscVal'] == 0)
df.loc[mask2, 'MiscFeature'] = 'Nmf'

#3.adım 
# 1. Adım: 'MiscFeature' ve 'MiscVal' değerlerini ayır
df_with_misc = df[df['MiscFeature'].notna()]
df_missing_misc = df[df['MiscFeature'].isna()]

# 2. Adım: Benzer 'MiscVal' değerlerini bulmak için Nearest Neighbors modelini oluştur
X = df_with_misc[['MiscVal']]
y = df_with_misc['MiscFeature']

# Nearest Neighbors modelini kur
nbrs = NearestNeighbors(n_neighbors=5, algorithm='auto').fit(X)

# Her eksik 'MiscFeature' için en yakın komşuları bul
for index, row in df_missing_misc.iterrows():
    value = row['MiscVal']
    distances, indices = nbrs.kneighbors([[value]])
    
    # Komşuların MiscFeature değerlerini al
    nearest_features = y.iloc[indices[0]].values
    mode_value = pd.Series(nearest_features).mode().iloc[0]
    
    # DataFrame'e doldur
    df.at[index, 'MiscFeature'] = mode_value
```

##### 3- handling with the Alley

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier

data= df[["Alley","LotShape",'Street','LotArea']]
# Kategorik verileri sayısal verilere dönüştürme
df_encoded = pd.get_dummies(data, columns=['Street', 'LotShape'])

# Boş 'Alley' değerlerini ve bunları dolduracak olanları ayır
df_notna = df_encoded[df_encoded['Alley'].notna()]
df_na = df_encoded[df_encoded['Alley'].isna()]

# Özellikler ve hedef değişken
X = df_notna.drop('Alley', axis=1)
y = df_notna['Alley']
X_na = df_na.drop('Alley', axis=1)

# Eğitim ve test setlerine ayırma (train/test)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# KNeighborsClassifier modelini kur
knn = KNeighborsClassifier(n_neighbors= 3)
knn.fit(X_train, y_train)

# Test setinde tahminler yapma
y_pred = knn.predict(X_test)

# Doğruluk skorunu hesaplama
accuracy = accuracy_score(y_test, y_pred)


```

```python
def fill_bsmt_null_with_mode(df):
    # Bsmt ile başlayan sütunları bul
    bsmt_columns = [col for col in df.columns if col.startswith('Bsmt')]
    
    # Her sütunun null değerlerini modları ile doldur
    for col in bsmt_columns:
        mode_value = df[col].mode()[0]  # Mod değerini hesapla
        df[col].fillna(mode_value, inplace=True)  # Null değerleri mod ile doldur
```
